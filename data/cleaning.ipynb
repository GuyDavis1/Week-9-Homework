```python
post=pd.read_csv("base_data_for_cleaning\postcode_suburb.csv")
post
```

```python
prop=prop.merge(post,left_on="Postcode",right_on="postcode")
prop=prop[(prop.state=="VIC")]
prop
```

```python
prop[prop.suburb=="BOX HILL NORTH"]
```

```python
prop=prop[["LGA","suburb"]].drop_duplicates()
prop.suburb=prop.suburb.str.title()
prop
```

```python
prop[prop.suburb.str.contains("Box")]
```

```python
suburb_rents=pd.read_csv("base_data_for_cleaning\suburb_rents.csv")
suburb_rents["Suburb"]=pd.DataFrame(suburb_rents.Suburb.str.split("-"))
suburb_rents=suburb_rents.explode(column="Suburb")
suburb_rents
```

```python
suburb_rents[suburb_rents.Suburb.str.contains("Box")]
```

```python
prop.drop_duplicates()
```

```python
merged=suburb_rents.merge(prop,left_on="Suburb",right_on="suburb",how="outer")

merged
```

```python
merged[merged.suburb=="Box Hill North"]
```

```python
merged.isnull().sum()
```

```python
lga_meds=merged.groupby("LGA").Median.mean().reset_index()
lga_meds.columns=["LGA","lga_med"]
lga_meds
```

```python
merged=merged.merge(lga_meds,how="left",on="LGA")
merged
```

```python
merged.iloc[merged.Median.isnull(),1]=merged[merged.Median.isnull()].lga_med
merged[merged.suburb=="Box Hill North"]
```

```python
merged=merged[["suburb","Median"]][~merged.suburb.isnull()].drop_duplicates()
merged
```

```python
merged[merged.suburb.str.contains("Box")]
```

```python
merged[["suburb","Median"]].to_csv("suburb_rents.csv")
```
