```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prop=pd.read_csv(r"C:\Users\Guy\Desktop\Monash uni\S6\FIT3179\vis 2/data/Melbourne_Housing_Full.csv")
prop=prop[~prop.CouncilArea.isnull()]
prop=prop[prop.CouncilArea.str.contains("City")]
prop
```

```python
suburb_rents=pd.read_csv("base_data_for_cleaning\suburb_rents.csv")
```

```python
suburb_rents=pd.read_csv("base_data_for_cleaning\suburb_rents.csv")
suburb_rents["Suburb"]=pd.DataFrame(suburb_rents.Suburb.str.split("-"))
suburb_rents=suburb_rents.explode(column="Suburb")
suburb_rents
```

```python
suburb_rents[suburb_rents.Suburb.str.contains("Box")]
```

```python
all_suburbs=prop.groupby(["Suburb","CouncilArea"]).Type.count().reset_index()
all_suburbs["Suburb"]=all_suburbs["Suburb"].str.title()
all_suburbs=all_suburbs[["Suburb","CouncilArea"]]
all_suburbs
```

```python
merged=all_suburbs.merge(suburb_rents,on="Suburb",how="outer")
merged
```

```python
merged.isnull().sum()
```

```python
lga_meds=merged.groupby("CouncilArea").Median.mean().reset_index()
lga_meds.columns=["CouncilArea","lga_med"]
```

```python
merged=merged.merge(lga_meds,how="left")
merged
```

```python
merged
```

```python
merged.iloc[merged.Median.isnull(),2]=merged[merged.Median.isnull()].lga_med
```

```python
merged[["Suburb","Median"]].to_csv("suburb_rents.csv")
```
